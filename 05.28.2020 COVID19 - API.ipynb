```python
df = pd.read_csv("https://covidtracking.com/api/v1/states/daily.csv")
df['Date'] = pd.to_datetime(df["dateChecked"])
df.sort_values(by="Date", ascending=True, inplace=True)
```

## Cumulative number of positive tests by state

```python
df['positive'] = df["positive"].astype(float)
```

```python
sns.set(rc={'figure.figsize': (18,18)})
sns.lineplot('Date','positive', hue = 'state', data = df, )
```

## Number of patients currently in ICU

```python

sns.set(rc={'figure.figsize': (18,18)})
sns.lineplot('Date','inIcuCurrently', hue = 'state', data = df, )
```
